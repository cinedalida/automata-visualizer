## CONTEXT FREE GRAMMAR (CFG)
### Problem 1: (1+0)* (11+00) (00+11)* (1+0+11) (1+0+11)* (101+111) (101+111)* (1+0*+11) (1+0*+11)*

```python
'''
FORMAT
Converted through on DFA

Problem 1: 
(1+0)* (11+00) (00+11)* (1+0+11) (1+0+11)* (101+111) (101+111)* (1+0*+11) (1+0*+11)*
'''

from collections import deque
import re

cfg = {
    "Q0": ["1Q1", "0Q2"],
    "Q1": ["0Q2", "1Q3"],
    "Q2": ["1Q1", "0Q3"],
    "Q3": ["0Q4", "1Q4"],
    "Q4": ["0Q3", "1Q5"],
    "Q5": ["0Q6", "1Q8"],
    "Q6": ["0Q4", "1Q9"],
    "Q7": ["0Q4", "1Q9"],
    "Q8": ["0Q7", "1Q9"],
    "Q9": ["0Q9", "1Q9", ""]
}

def get_leftmost_variable(s):
    match = re.search(r'Q\d+', s)
    return (match.group(), match.start()) if match else (None, -1)

def bfs(target, max_depth=50):
    queue = deque([("Q0", [])])

    while queue:
        current, history = queue.popleft()

        derived = "".join(c for c in current if c in "01")

        # prefix pruning
        if not target.startswith(derived):
            continue

        # full match
        if derived == target and not re.search(r'Q\d+', current):
            return {
                "accepted": True,
                "sequence": history + [current]
            }

        if len(history) > max_depth:
            continue

        var, i = get_leftmost_variable(current)

        if var:
            for prod in cfg[var]:
                new = current[:i] + prod + current[i+len(var):]
                queue.append((new, history + [current]))

    return {
        "accepted": False,
        "sequence": []
    }

# RUN
user_input = input("Enter string: ")

result = bfs(user_input)

print("\nAccepted:", result["accepted"])

if result["sequence"]:
    print("\nDerivation:")
    for i, step in enumerate(result["sequence"]):
        print(i, ":", step)
else:
    print("\nNo derivation found.")
```

## CONTEXT FREE GRAMMAR (CFG)
### Problem 2: (bab)* (b+a) (bab+aba) (a+b)* (aa+bb)* (b+a+bb) (a+b)* (aa+bb)

```python
'''
FORMAT
Converted through on DFA

Problem 2: 
(bab)* (b+a) (bab+aba) (a+b)* (aa+bb)* (b+a+bb) (a+b)* (aa+bb)
'''


from collections import deque
import re

cfg = {
    "q0": ["aq2", "bq1"],

    "q1": ["aq4", "bq3"],
    "q2": ["bq3", "aq5"],

    "q3": ["aq7", "bT"],
    "q4": ["aq5", "bq6"],
    "q5": ["aT", "bq8"],

    "q6": ["aq10", "bq1"],
    "q7": ["bq9", "aT"],

    "q8": ["aq10", "bT"],
    "q9": ["aq11", "bq11"],

    "q10": ["aq11", "bq11"],

    "q11": ["bq12", "aq13"],
    "q12": ["bq14", "aq13"],
    "q13": ["bq12", "aq15"],

    # final states (ε production)
    "q14": ["aq13", "bq14", ""],
    "q15": ["aq15", "bq12", ""],

    # trap state
    "T": ["aT", "bT"]
}

def get_leftmost_variable(s):
    match = re.search(r'q\d+|T', s)
    return (match.group(), match.start()) if match else (None, -1)


def bfs(target, max_depth=50):
    queue = deque([("q0", [])])

    while queue:
        current, history = queue.popleft()

        # extract terminals only
        derived = "".join(c for c in current if c in "ab")

        # pruning: if mismatch, stop exploring
        if not target.startswith(derived):
            continue

        # check acceptance
        if derived == target and not re.search(r'q\d+|T', current):
            return {
                "accepted": True,
                "sequence": history + [current]
            }

        if len(history) > max_depth:
            continue

        var, i = get_leftmost_variable(current)

        if var:
            for prod in cfg[var]:
                new = current[:i] + prod + current[i+len(var):]
                queue.append((new, history + [current]))

    return {
        "accepted": False,
        "sequence": []
    }


if __name__ == "__main__":
    user_input = input("Enter string (a/b only): ")

    result = bfs(user_input)

    print("\nAccepted:", result["accepted"])

    if result["sequence"]:
        print("\nDerivation Steps:")
        for i, step in enumerate(result["sequence"]):
            print(f"{i}: {step}")
    else:
        print("\nNo derivation found.")

```

## PUSH DOWN AUTOMATA (PDA)
### Problem 1: (1+0)* (11+00) (00+11)* (1+0+11) (1+0+11)* (101+111) (101+111)* (1+0*+11) (1+0*+11)*

```python
# =========================
# PDA TRANSITIONS (from CFG)
# =========================
pda_transitions = {
    ("Q0", "1", "Z0"): ("Q1", "Z0"),
    ("Q0", "0", "Z0"): ("Q2", "Z0"),

    ("Q1", "0", "Z0"): ("Q2", "Z0"),
    ("Q1", "1", "Z0"): ("Q3", "Z0"),

    ("Q2", "1", "Z0"): ("Q1", "Z0"),
    ("Q2", "0", "Z0"): ("Q3", "Z0"),

    ("Q3", "0", "Z0"): ("Q4", "Z0"),
    ("Q3", "1", "Z0"): ("Q4", "Z0"),

    ("Q4", "0", "Z0"): ("Q3", "Z0"),
    ("Q4", "1", "Z0"): ("Q5", "Z0"),

    ("Q5", "0", "Z0"): ("Q6", "Z0"),
    ("Q5", "1", "Z0"): ("Q8", "Z0"),

    ("Q6", "0", "Z0"): ("Q4", "Z0"),
    ("Q6", "1", "Z0"): ("Q9", "Z0"),

    ("Q7", "0", "Z0"): ("Q4", "Z0"),
    ("Q7", "1", "Z0"): ("Q9", "Z0"),

    ("Q8", "0", "Z0"): ("Q7", "Z0"),
    ("Q8", "1", "Z0"): ("Q9", "Z0"),

    ("Q9", "0", "Z0"): ("Q9", "Z0"),
    ("Q9", "1", "Z0"): ("Q9", "Z0"),
}

# =========================
# FINAL STATES (from CFG)
# =========================
final_states = {"Q9"}

# =========================
# PDA SIMULATOR
# =========================
def run_pda(input_string):
    state = "Q0"
    stack = ["Z0"]

    for symbol in input_string:
        top = stack[-1]
        key = (state, symbol, top)

        if key not in pda_transitions:
            return False, state

        new_state, new_stack_top = pda_transitions[key]

        state = new_state
        stack[-1] = new_stack_top

    return state in final_states, state


# =========================
# TEST LOOP
# =========================
if __name__ == "__main__":
    print("PDA Simulator (from CFG)\nType 'exit' to stop")

    while True:
        s = input("\nEnter binary string (0/1): ")

        if s == "exit":
            break

        accepted, final_state = run_pda(s)

        print("Accepted:", accepted)
        print("Final State:", final_state)
```

## PUSH DOWN AUTOMATA (PDA)
### Problem 2: (bab)* (b+a) (bab+aba) (a+b)* (aa+bb)* (b+a+bb) (a+b)* (aa+bb)

```python
pda_transitions = {
    ("q0", "a", "Z0"): ("q2", "Z0"),
    ("q0", "b", "Z0"): ("q1", "Z0"),

    ("q1", "a", "Z0"): ("q4", "Z0"),
    ("q1", "b", "Z0"): ("q3", "Z0"),

    ("q2", "b", "Z0"): ("q3", "Z0"),
    ("q2", "a", "Z0"): ("q5", "Z0"),

    ("q3", "a", "Z0"): ("q7", "Z0"),
    ("q3", "b", "Z0"): ("T", "Z0"),

    ("q4", "a", "Z0"): ("q5", "Z0"),
    ("q4", "b", "Z0"): ("q6", "Z0"),

    ("q5", "a", "Z0"): ("T", "Z0"),
    ("q5", "b", "Z0"): ("q8", "Z0"),

    ("q6", "a", "Z0"): ("q10", "Z0"),
    ("q6", "b", "Z0"): ("q1", "Z0"),

    ("q7", "b", "Z0"): ("q9", "Z0"),
    ("q7", "a", "Z0"): ("T", "Z0"),

    ("q8", "a", "Z0"): ("q10", "Z0"),
    ("q8", "b", "Z0"): ("T", "Z0"),

    ("q9", "a", "Z0"): ("q11", "Z0"),
    ("q9", "b", "Z0"): ("q11", "Z0"),

    ("q10", "a", "Z0"): ("q11", "Z0"),
    ("q10", "b", "Z0"): ("q11", "Z0"),

    ("q11", "b", "Z0"): ("q12", "Z0"),
    ("q11", "a", "Z0"): ("q13", "Z0"),

    ("q12", "b", "Z0"): ("q14", "Z0"),
    ("q12", "a", "Z0"): ("q13", "Z0"),

    ("q13", "b", "Z0"): ("q12", "Z0"),
    ("q13", "a", "Z0"): ("q15", "Z0"),

    ("q14", "a", "Z0"): ("q13", "Z0"),
    ("q14", "b", "Z0"): ("q14", "Z0"),

    ("q15", "a", "Z0"): ("q15", "Z0"),
    ("q15", "b", "Z0"): ("q12", "Z0"),

    ("T", "a", "Z0"): ("T", "Z0"),
    ("T", "b", "Z0"): ("T", "Z0"),
}

# =========================
# PDA Simulator
# =========================
def run_pda(input_string):
    state = "q0"
    stack = ["Z0"]

    for symbol in input_string:
        top = stack[-1]

        key = (state, symbol, top)

        if key not in pda_transitions:
            return False, state

        new_state, new_stack_top = pda_transitions[key]

        # update state
        state = new_state

        # stack (trivial PDA → we just replace top)
        stack[-1] = new_stack_top

    # Accepting states
    if state in {"q14", "q15"}:
        return True, state
    else:
        return False, state


# =========================
# RUN TESTS
# =========================
while True:
    s = input("\nEnter string (a/b only, or 'exit'): ")

    if s == "exit":
        break

    accepted, final_state = run_pda(s)

    print("Accepted:", accepted)
    print("Final State:", final_state)
```
